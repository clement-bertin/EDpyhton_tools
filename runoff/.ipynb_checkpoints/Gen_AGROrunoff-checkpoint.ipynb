### a. Read raw AGRO Mackenzie discharge data

```python
Disch_df = Raw_df.interpolate(method='quadratic')
Disch_df.to_csv(DatIn_pth+'Mackenzie_Q_reproc.csv', sep=';')
```

```python
plt.figure(figsize=(15,4))
### Plot
plt.plot(Disch_df.discharge)
### Params
plt.grid(True)
plt.xlim(Date[0],Date[-1])
plt.tick_params(labelsize=14)
plt.show()
```

Create a plottable climatology matrix

```python
FWclim_plt = np.zeros(len(Disch_df))
itr = 0
for yr in list(set(Date.year)):
    if len(np.where(Date.year==yr)[0]) == 366:
        FWclim_plt[itr:itr+366] = Disch_Clim
        itr += 366
    else:
        FWclim_plt[itr:itr+365] = np.concatenate((Disch_Clim[:59],Disch_Clim[60:]))
        itr += 365
```

```python
def extend_clim(Adf, Cdf):
    # Create final array
    Farray = np.zeros(len(Adf))
    # Get years
    Yrs = np.unique(Adf.index.year.values)
    # Loop on the years 
    itr = 0
    for yr in Yrs:
        if len(Adf[Adf.index.year==yr]) == 366:
            Farray[itr:itr+366] = Cdf
            itr += 366
        else:
            Farray[itr:itr+365] = np.concatenate((Cdf[:59],Cdf[60:]))
            itr += 365
    Fdf = pd.DataFrame(data=Farray, index=Adf.index, columns=['CO2f 3qt'])
    return(Fdf)
```

## V. Process the mac270 Freshwater AGRO internual forcing

###### Create the FW forcing matrix

$r[m.s^{-1}] = Q[m^3.s^{-1}]*W$ 

where $W = \frac{w}{A.N}$ with:
- $w$= Morley (2012) outlet weight
- A = cell area
- N = Number of entry cells for the outlet

```python
# Create the W matrix
id_inlet = np.where(inlet>0)
N = np.ones(inlet.shape)
for i in range(int(inlet.max())):
    N[np.where(inlet==i+1)] = list(inlet[id_inlet]).count(i+1)
W = w/(Area*N)
W_exp = np.repeat(np.expand_dims(W,axis=0), len(Disch_df),axis=0)
```

```python
# Compute the runoff matrix
FWmac270 = np.zeros((len(Disch_df),XC.shape[0],YC.shape[1]))
for j in range(len(id_inlet[0])):
    FWmac270[:,id_inlet[0][j],id_inlet[1][j]] = W_exp[:,id_inlet[0][j],id_inlet[1][j]]*Disch_df.discharge.values
```

###### Verification plot

```python
plt.figure(figsize=(15,4))
for i in range(len(np.where(inlet>0)[0])):
    plt.plot(Date,FWmac270[:,np.where(inlet>0)[0][i],np.where(inlet>0)[1][i]])
plt.grid(True)
plt.xlim([Date[0],Date[-1]])
plt.show()
```

## V. Process the mac270 Freshwater AGRO climatological forcing

###### Create climatolgy FW forcing matrix

```python
# Create the W matrix for climatology
W_clim = np.repeat(np.expand_dims(W,axis=0), len(Disch_Clim),axis=0)
```

```python
# Bissextile year
FWmac270_climBx = np.zeros((len(Disch_Clim),XC.shape[0],YC.shape[1]))
for j in range(len(id_inlet[0])):
    FWmac270_climBx[:,id_inlet[0][j],
                    id_inlet[1][j]] = W_clim[:,id_inlet[0][j],id_inlet[1][j]]*Disch_Clim
```

```python
# Normal year
FWmac270_climNm = np.zeros((len(Disch_Clim)-1,XC.shape[0],YC.shape[1]))
for j in range(len(id_inlet[0])):
    FWmac270_climNm[:,id_inlet[0][j],
                    id_inlet[1][j]] = W_clim[:-1,id_inlet[0][j],
                                             id_inlet[1][j]]*np.concatenate((Disch_Clim[:59],Disch_Clim[60:]))
```

###### Verification plot

```python
plt.figure(figsize=(15,4))
for i in range(len(np.where(inlet>0)[0])):
    plt.plot(FWmac270_climNm[:,np.where(inlet>0)[0][i],np.where(inlet>0)[1][i]])
plt.grid(True)
plt.show()
```

## VII. Save Forcing matrixes

```python
def bin_save(fnm, array):
    newFile = open(fnm, "wb")
    bitearray = pack(">%sf" % len(array.flatten()),*array.flatten())
    newFile.write(bitearray);
```

### a. Save interanual forcing matrix

```python
if not 'AGRO_interan' in os.listdir(Out_pth):
    os.mkdir(Out_pth+"AGRO_interan/")
# Save forcing file
for yr in list(set(Date.year)):
    bin_save(Out_pth+"AGRO_interan/AGRO_runoff_mac270_"+str(yr),FWmac270[np.where(Date.year==yr)])
```

### b. Save climatology forcing matrix

```python
if not 'AGRO_clim' in os.listdir(Out_pth):
    os.mkdir(Out_pth+"AGRO_clim/")
# Save forcing file
for yr in list(set(Date.year)):
    if len(np.where(Date.year==yr)[0]) == 366:
        bin_save(Out_pth+"AGRO_clim/AGRO_runoff_mac270_"+str(yr),FWmac270_climBx)
    else:
        bin_save(Out_pth+"AGRO_clim/AGRO_runoff_mac270_"+str(yr),FWmac270_climNm)
```
